# Lambda Calculus

```haskell
import Prelude
```

## 1. Defining Booleans

```haskell
true = \x -> \y -> x
false = \x -> \y -> y
```

## 1.1 AND

```haskell
and' = \x -> \y -> x y false
```

```haskell
toBool bool = bool "True" "False"
```

```haskell
toBool true
```

```haskell
toBool false
```

```haskell
toBool $ and' true false
```

```haskell
toBool $ and' true true
```

```haskell
toBool $ and' false false
```

```haskell
toBool $ and' false true
```

## 1.2 OR

```haskell
or' = \x y -> x true y
```

```haskell
toBool $ or' false false
```

```haskell
toBool $ or' false true
```

```haskell
toBool $ or' true true
```

```haskell
toBool $ or' true false
```

## 1.3 NOT

```haskell
not' = \x -> x false true
```

```haskell
toBool $ not' true
```

```haskell
toBool $ not' false
```
